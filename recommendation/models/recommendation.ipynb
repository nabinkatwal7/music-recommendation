```python
import pandas as pd 
from sklearn.neighbors import NearestNeighbors
import pickle 
```

```python
spotify_df = pd.read_csv('./data/spotify_audio_features.csv')
spotify_df
```

```python
spotify_df.info()
```

```python
model_neighbor = NearestNeighbors(n_neighbors=8, metric = 'cosine', algorithm = 'auto')
model_neighbor.fit(spotify_df.iloc[:,3:])
```

```python
#Inference
distances, indices = model_neighbor.kneighbors(spotify_df.iloc[1, 3:].values.reshape(1,-1), n_neighbors=7)
distances, indices 
```

```python
spotify_df.iloc[1,:2].tolist()
```

```python
for i in range(0, len(distances.flatten())): 
    if i==0:
        print(f'Recommendations for {spotify_df.index[1]}: \n')
    else:
        print(f'{i}: {spotify_df.iloc[indices.flatten()[i],:2].tolist()}, with distance of {distances.flatten()[i]}')
        
```

```python
with open('./models/spotify-recommender.pkl','wb') as f:
    pickle.dump(model_neighbor, f)
```
